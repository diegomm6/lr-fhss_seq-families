# DYNAMIC DOPPLER EXHAUSTIVE SEARCH

```python
data = pd.read_csv('dyndoppler/500t-6tg-25fg-srch.csv', index_col=0, header=None).T
nodes = data['nodes'].to_numpy()

decoded_decodable = np.add(data['nodrop-cr1-500p-decoded_hrd_pld'].to_numpy(), data['nodrop-cr1-500p-decodable_pld'].to_numpy())
fhslocator_results = np.add(data['nodrop-cr1-500p-tp'].to_numpy(), data['nodrop-cr1-500p-fp'].to_numpy())

i=0
while fhslocator_results[i] < nodes[i]*1.1:
    i+=1

lim = nodes[i]
for j in range(i, len(nodes)):
    decoded_decodable[j]=0
    

plt.figure(figsize=(10,10))

plt.plot(nodes, data['nodrop-cr1-500p-decoded_hrd_pld'], label=f'decoded_hrd_pld')
plt.plot(nodes, data['nodrop-cr1-500p-decoded_hdr'], label=f'decoded_hdr')
plt.plot(nodes, data['nodrop-cr1-500p-decodable_pld'], label=f'decodable_pld')
plt.plot(nodes, data['nodrop-cr1-500p-collided_hdr_pld'], label=f'collided_hdr_pld')

plt.plot(nodes, nodes, label=f'x=y', color='k', linestyle='dotted')

plt.title('cr1, 8-30 fragments')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('count')
plt.xlim(1e1,1e3)
plt.ylim(0,1.5e2)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.plot(nodes, fhslocator_results, label=f'tp+fp', color='r', linestyle='dashed')
plt.plot(nodes, data['nodrop-cr1-500p-tp'], label=f'tp')
plt.plot(nodes, data['nodrop-cr1-500p-fp'], label=f'fp')
plt.plot(nodes, data['nodrop-cr1-500p-fn'], label=f'fn')
plt.plot(nodes, nodes, label=f'x=y', color='k', linestyle='dotted')

plt.axvline(x=lim, color='m', label=f'limit={lim}')

plt.title('cr1, 8-30 fragments, 10 runs')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('count')
plt.xlim(1e1,1e3)
plt.ylim(0,1e3)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.plot(nodes, data['nodrop-cr1-500p-decoded_hrd_pld'], label=f'decoded_hrd_pld')
plt.plot(nodes, data['nodrop-cr1-500p-decoded_hdr'], label=f'decoded_hdr')
plt.plot(nodes, data['nodrop-cr1-500p-decodable_pld'], label=f'decodable_pld')
plt.plot(nodes, data['nodrop-cr1-500p-collided_hdr_pld'], label=f'collided_hdr_pld')
plt.plot(nodes, nodes, label=f'x=y', color='k', linestyle='dotted')

plt.plot(nodes, decoded_decodable, label=f'payloads with FHS locator')

plt.title('cr1, 8-30 fragments')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('count')
plt.xlim(1e1,1e3)
plt.ylim(0,1.5e2)
plt.grid()
plt.legend()
plt.show()
```
