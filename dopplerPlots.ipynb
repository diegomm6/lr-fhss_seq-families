```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# no header drop

```python
data = pd.read_csv('doppler/cr2-7000t-6tg-7fg.csv', index_col=0, header=None).T
nodes = data['nodes'].to_numpy()

plt.figure(figsize=(10,8))

plt.plot(nodes, data['nodrop-1000p-decoded_hrd_pld'], label=f'decoded_hrd_pld')
plt.plot(nodes, data['nodrop-1000p-decoded_hdr'], label=f'decoded_hdr')
plt.plot(nodes, data['nodrop-1000p-decodable_pld'], label=f'decodable_pld')
plt.plot(nodes, data['nodrop-1000p-collided_hdr_pld'], label=f'collided_hdr_pld')

plt.plot(nodes, nodes, label=f'x=y', color='k', linestyle='dotted')

plt.title('cr2, 8-30 fragments')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('count')
plt.xlim(1e2,1e4)
plt.ylim(0,2e3)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.plot(nodes, np.divide(data['nodrop-1000p-decoded_hrd_pld'], nodes), label=f'decoded_hrd_pld')
plt.plot(nodes, np.divide(data['nodrop-1000p-decoded_hdr'], nodes), label=f'decoded_hdr')
plt.plot(nodes, np.divide(data['nodrop-1000p-decodable_pld'], nodes), label=f'decodable_pld')
plt.plot(nodes, np.divide(data['nodrop-1000p-collided_hdr_pld'], nodes), label=f'collided_hdr_pld')

plt.title('cr2, 8-30 fragments')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('rate')
plt.xlim(1e2,1e4)
plt.grid()
plt.legend()
plt.show()
```
