```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('sinrdecoder/sinr.csv', index_col=0, header=None).T
nodes = data['nodes'].to_numpy()

fhslocator_results = np.add(data['driver-tp'].to_numpy(), data['driver-fp'].to_numpy())
decoded_decodable = np.add(data['driver-decoded_hrd_pld'].to_numpy(), data['driver-decodable_pld'].to_numpy())
decodable_collided = np.add(data['driver-decodable_pld'].to_numpy(), data['driver-collided_hdr_pld'].to_numpy())

plt.figure(figsize=(10,10))

plt.plot(nodes, data['driver-decoded_hrd_pld'], label=f'decoded_hrd_pld (n1)')
plt.plot(nodes, data['driver-decoded_hdr'], label=f'decoded_hdr (n2)')
plt.plot(nodes, data['driver-decodable_pld'], label=f'decodable_pld (n3)')
plt.plot(nodes, data['driver-collided_hdr_pld'], label=f'collided_hdr_pld (n4)')

plt.plot(nodes, nodes, label=f'total txs (N)', color='k', linestyle='dotted')

plt.title('LR-FHSS Decoder results')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('count')
#plt.xlim(1e1,1e3)
#plt.ylim(0,1.1e2)
plt.grid()
plt.legend()
plt.show()
```

```python
data = pd.read_csv('nodoppler/nodop.csv', index_col=0, header=None).T
nodes = data['nodes'].to_numpy()

fhslocator_results = np.add(data['driver-tp'].to_numpy(), data['driver-fp'].to_numpy())
decoded_decodable = np.add(data['driver-decoded_hrd_pld'].to_numpy(), data['driver-decodable_pld'].to_numpy())
decodable_collided = np.add(data['driver-decodable_pld'].to_numpy(), data['driver-collided_hdr_pld'].to_numpy())

plt.figure(figsize=(10,8))

plt.plot(nodes, data['driver-decoded_hrd_pld'], label=f'decoded_hrd_pld (n1)', linewidth=2)
plt.plot(nodes, data['driver-decoded_hdr'], label=f'decoded_hdr (n2)', linewidth=2)
plt.plot(nodes, data['driver-decodable_pld'], label=f'decodable_pld (n3)', linewidth=2)
plt.plot(nodes, data['driver-collided_hdr_pld'], label=f'collided_hdr_pld (n4)', linewidth=2)
plt.plot(nodes, decoded_decodable, label=f'decoded + decodable (n1+n3)', linewidth=2)

plt.plot(nodes, nodes, label=f'total txs (N)', color='k', linestyle='dotted', linewidth=2)

plt.title('LR-FHSS Decoder results (terrestrial)', fontsize=18)
plt.xlabel('transmissions', fontsize=18)
plt.ylabel('count', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xscale('log')
#plt.yscale('log')
plt.xlim(1e1,1e3)
plt.ylim(0,2.5e2)
plt.grid()
plt.legend(fontsize=14)
plt.savefig("statusresultsnodop.pdf")
```

```python
plt.figure(figsize=(10,5))

plt.plot(nodes, fhslocator_results, label=f'tp + fp (Ĉ)', color='r', linestyle='dashed', linewidth=2)
plt.plot(nodes, data['driver-tp'], label=f'true positives (tp)', linewidth=2)
plt.plot(nodes, data['driver-fp'], label=f'false positives (fp)', linewidth=2)
plt.plot(nodes, data['driver-fn'], label=f'false negatives (fn)', linewidth=2)
plt.plot(nodes, decodable_collided, label=f'collided hdrs (C)', color='k', linestyle='dotted', linewidth=2)

#plt.axvline(x=lim, color='m', label=f'limit={lim}')

plt.title('FHS Locator results (terrestrial)', fontsize=18)
plt.xlabel('transmissions', fontsize=18)
plt.ylabel('count', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e1,1e3)
#plt.ylim(0,1e3)
plt.grid()
plt.legend(fontsize=14)
plt.savefig("fhslresultsnodop.pdf", bbox_inches='tight')
```

```python
plt.figure(figsize=(10,5))

plt.plot(nodes, data['driver-chocc'], label='channel occupancy [%]', linewidth=2)
plt.plot(nodes, np.divide(decoded_decodable, nodes), label='decodable pld rate = (n1+n3)/N', linewidth=2)
plt.plot(nodes, np.divide(decodable_collided, fhslocator_results), label='C / Ĉ = (n3+n4)/(tp+fp)', linewidth=2)

plt.title('OCW channel occupancy (terrestrial)', fontsize=18)
plt.xlabel('transmissions', fontsize=18)
plt.ylabel('rate', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xscale('log')
plt.xlim(1e1,1e3)
plt.grid()
plt.legend(fontsize=14)
plt.savefig("ocwchoccnodop.pdf", bbox_inches='tight')
```

```python
precsision = np.divide(data['driver-tp'], np.add(data['driver-tp'], data['driver-fp']))
recall = np.divide(data['driver-tp'], np.add(data['driver-tp'], data['driver-fn']))
f1score = 2 * np.divide(np.multiply(precsision, recall), np.add(precsision, recall))

plt.figure(figsize=(10,4))

plt.plot(nodes, f1score, linewidth=2)

plt.title('FHS Locator F1 score (terrestrial)', fontsize=18)
plt.xlabel('transmissions', fontsize=18)
plt.ylabel('f1', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xscale('log')
plt.xlim(1e1,1e3)
plt.grid()
plt.legend(fontsize=14)
plt.savefig("f1scorenodop.pdf", bbox_inches='tight')
```
