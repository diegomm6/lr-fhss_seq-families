```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('sinrdecoder/sinr.csv', index_col=0, header=None).T
nodes = data['nodes'].to_numpy()

fhslocator_results = np.add(data['driver-tp'].to_numpy(), data['driver-fp'].to_numpy())
decoded_decodable = np.add(data['driver-decoded_hrd_pld'].to_numpy(), data['driver-decodable_pld'].to_numpy())
decodable_collided = np.add(data['driver-decodable_pld'].to_numpy(), data['driver-collided_hdr_pld'].to_numpy())

plt.figure(figsize=(10,10))

plt.plot(nodes, data['driver-decoded_hrd_pld'], label=f'decoded_hrd_pld (n1)')
plt.plot(nodes, data['driver-decoded_hdr'], label=f'decoded_hdr (n2)')
plt.plot(nodes, data['driver-decodable_pld'], label=f'decodable_pld (n3)')
plt.plot(nodes, data['driver-collided_hdr_pld'], label=f'collided_hdr_pld (n4)')

plt.plot(nodes, nodes, label=f'total txs (N)', color='k', linestyle='dotted')

plt.title('LR-FHSS Decoder results')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('count')
#plt.xlim(1e1,1e3)
#plt.ylim(0,1.1e2)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(nodes, data['driver-chocc'], label='channel occupancy [%]')
plt.plot(nodes, np.divide(decoded_decodable, nodes), label='decodable pld rate = (n1+n3)/N')
plt.plot(nodes, np.divide(decodable_collided, fhslocator_results), label='C / Ĉ = (n3+n4)/(tp+fp)')

plt.title('OCW channel occuapncy')
plt.xlabel('nodes')
plt.xscale('log')
plt.ylabel('rate')
plt.xlim(1e1,1e3)
plt.grid()
plt.legend()
plt.show()
```
